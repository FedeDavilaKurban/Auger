```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np

filename_g = '../data/VLS_ang5_cz_700control_def.txt'

data = ascii.read(filename_g)

plt.hist(data['deflection'], bins=15, histtype='step', label='deflection')
plt.axvline(x=np.median(data['deflection']), color='k', linestyle='--', label='mean')
plt.axvline(x=1, color='k', linestyle='--', label='mean')
plt.show()
```
