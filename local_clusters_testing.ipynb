```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np
from astropy.coordinates import SkyCoord
import astropy.units as u

# Read data
clusters = ascii.read("../data/Local_Clusters_z0.055.txt")

# Create SkyCoord object from RA and Dec (assumed columns "RA" and "Dec" in degrees)
coord = SkyCoord(ra=clusters["RAJ2000"], dec=clusters["DEJ2000"], unit=u.deg, frame='icrs')

# Transform to galactic coordinates
galactic = coord.galactic
gal_lat = galactic.b.deg

# Filter out clusters with dec > 45
mask_dec = clusters["DEJ2000"] <= 45

# Filter out clusters with |b| < 5° (inside the plane)
mask = np.abs(gal_lat) >= 5
clusters_filtered = clusters[mask]

# Speed of light
c = 299792.458  # km/s

# Print number of clusters before and after filtering for redshift velocity between 1200 and 4000 km/s
clusters_filtered = clusters_filtered[(clusters_filtered["z"] * c >= 1200) & (clusters_filtered["z"] * c <= 4000)]

print(f"Total clusters: {len(clusters)}")
print(f"Clusters after filtering: {len(clusters_filtered)}")


# Histogram of redshift velocity for filtered clusters
# plt.hist(clusters_filtered["z"] * c, bins=20)
# plt.axvline(1200, linestyle='dashed', c='k')
# plt.axvline(4000, linestyle='dashed', c='k')
# plt.xlabel("Velocity (km/s)")
# plt.ylabel("Number of clusters")
# plt.show()
```
