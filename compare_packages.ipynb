#### Codigo para plottear los resultados de los 3 paquetes que estoy probando

```python
import numpy as np
import matplotlib.pyplot as plt

fname_nbodykit = '../data/nbodykit.npz'
fname_treecorr = '../data/treecorr.npz'
fname_corrfunc = '../data/corrfunc.npz'

packages = ['nbodykit','treecorr','corrfunc']

for package in packages:
    fname = f'../data/{package}.npz'
    f = np.load(fname)
    x,y,e = f['arr_0'],f['arr_1'],f['arr_2']
    plt.errorbar(x,y,yerr=e,label=package)
plt.xlabel(r'$\theta(deg)$')
plt.ylabel(r'$\omega(\theta)$')
plt.xlim([10,90])
plt.legend()
plt.show
plt.savefig('../plots/auto_comparison.png')
```

#### Cross

```python
import numpy as np
import matplotlib.pyplot as plt

packages = ['nbodykit','treecorr','corrfunc']
lss = ['-', '--', ':']
nmult = 50

for gtype, c in zip(['bright','faint'],['b','r']):
    for package,ls in zip(packages,lss):
        fname = f'../data/{package}_cross_{gtype}.npz'
        print(fname)
        f = np.load(fname)
        x,y,e = f['arr_0'],f['arr_1'],f['arr_2']
        plt.errorbar(x,y,yerr=e,label=package+' - '+gtype,color=c,ls=ls)

plt.hlines(0,0,100,color='k')

plt.xlabel(r'$\theta(deg)$')
plt.ylabel(r'$\omega(\theta)$')
plt.xlim([10,90])
plt.legend()
plt.show
#plt.savefig('../plots/auto_comparison.png')
```
