```python
import numpy as np
from matplotlib import pyplot as plt
from astropy.io import ascii

gxs = ascii.read('../data/2MRSxWISE_sinBPTAGNs.txt')
print(gxs.columns)
gxs = gxs[gxs['K_abs'] < -24.0]
gxs = gxs[gxs['_DEJ2000'] < -20.0]

gxs_z8000_9400 = gxs[(gxs['cz'] > 8000)&(gxs['cz'] < 9400)]
gxs_z1200_3600 = gxs[(gxs['cz'] < 3600)&(gxs['cz'] > 1200)]
gxs = gxs[gxs['class']!= 1] 

auger = ascii.read('../data/Auger/events_a8_lb.dat')
auger = auger[auger['dec'] < -20.0]

#plt.hist(gxs['_DEJ2000'], bins=50, alpha=.5, label='Gxs', density=True)
plt.hist(gxs_z8000_9400['_DEJ2000'], bins=50, alpha=.5, label='Gxs_far', density=True)
plt.hist(gxs_z1200_3600['_DEJ2000'], bins=50, alpha=.5, label='Gxs_close', density=True)

plt.hist(auger['dec'], bins=50, alpha=.5, label='Auger', density=True)
plt.xlabel('Dec')
plt.legend()
plt.show()
```
